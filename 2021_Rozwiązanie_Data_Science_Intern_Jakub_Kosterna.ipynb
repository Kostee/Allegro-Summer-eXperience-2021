#### A

```python
def findA(df, favCats, location):
    df = df[df["city"] == location[0]]
    df = df.loc[df['product_category_name'].isin(favCats)]
    df = df.groupby(["product_id"]).count().reset_index()
    result_row = df.loc[df['city'].idxmax()]
    return result_row.product_id
```

#### B

In order to calculate it, we need function for finding 5 nearest cities, in which at least one product of verified category was sold.

```python
geolocation_unique_cities = geolocation_unique_zip_codes.drop(["geolocation_zip_code_prefix"], axis=1)
geolocation_unique_cities = geolocation_unique_cities.groupby(["geolocation_state",
                                                               "geolocation_city"]).mean().reset_index()
geolocation_unique_cities.columns = ["state", "city", "lat", "lng"]
geolocation_unique_cities.sample(5)
```

```python
def distance(lat1, lon1, lat2, lon2):
    R = 6373.0 # approximate radius of earth in km
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c
```

Example usage:

```python
print(distance(52.2296756, 21.0122287, 52.406374, 16.9251681))
```

```python
def getCitiesContainingCategorySell(dfProducts, categoryName):
    df_cities_category_sold = dfProducts[dfProducts["product_category_name"] == categoryName]
    cities_selling = df_cities_category_sold[["city"]].drop_duplicates()
    return list(cities_selling["city"])
```

Example usage:

```python
getCitiesContainingCategorySell(dfLogged, "utilidades_domesticas")[0:10]
len(getCitiesContainingCategorySell(dfLogged, "utilidades_domesticas"))
```

```python
def find5nearest(dfProducts, dfLocations, cityName, categoryName):
    cityRow = dfLocations[dfLocations['city'] == cityName].iloc[0]
    ourLat = cityRow.lat
    ourLng = cityRow.lng
    
    citiesContainingCat = getCitiesContainingCategorySell(dfProducts, categoryName)
    dfLocations = dfLocations[dfLocations["city"].isin(citiesContainingCat)]
    
    dfLocations["distance"] = dfLocations.apply(lambda x: distance(ourLat, ourLng, x["lat"], x["lng"]), axis=1)
    dfLocations = dfLocations.sort_values("distance")
    dfLocations = dfLocations[dfLocations["distance"] > 0]
    dfLocations = dfLocations.head(5)
    return list(dfLocations["city"])
```

Example usage:

```python
pd.options.mode.chained_assignment = None

find5nearest(dfLogged, geolocation_unique_cities, "guarulhos", "utilidades_domesticas")[0:10]
len(find5nearest(dfLogged, geolocation_unique_cities, "guarulhos", "utilidades_domesticas"))
```

```python
def findB(df, favCats, location):
    result = list()
    for cat in favCats:
        nearestCities = find5nearest(dfLogged, geolocation_unique_cities, location[0], cat)
        dfTmp = df[df["state"] == location[1]]
        dfTmp = df[~df["city"].isin(nearestCities)]
        dfTmp = dfTmp.loc[dfTmp['product_category_name'] == cat]
        dfTmp = dfTmp.groupby(["product_id"]).count().reset_index()
        result_row = dfTmp.loc[dfTmp['city'].idxmax()]
        result.append(result_row.product_id)
    return result
```

Example usage:

```python
findB(dfLogged, ["esporte_lazer", "brinquedos", "bebes"], ["tupa", "SP"])
```

#### C

```python
def findC(df, favCats, location):
    result = list()
    for cat in favCats:
        nearestCities = find5nearest(dfLogged, geolocation_unique_cities, location[0], cat)
        dfTmp = df[df["city"].isin(nearestCities)]
        dfTmp = dfTmp.loc[dfTmp['product_category_name'] == cat]
        dfTmp = dfTmp.groupby(["product_id"]).count().reset_index()
        result_row = dfTmp.loc[dfTmp['city'].idxmax()]
        result.append(result_row.product_id)
    return result
```

```python
findC(dfLogged, ["esporte_lazer", "brinquedos", "bebes"], ["tupa", "SP"])
```

#### D

```python
def findD(df, favCats, location):
    anyNone = False
    result = list()
    for cat in favCats:
        dfTmp = df[df["city"] == location[0]]
        dfTmp = df[df["product_category_name"] == cat]
        if len(dfTmp) > 0:
            dfTmp = dfTmp.groupby(["product_id"]).count().reset_index()
            result_row = dfTmp.loc[dfTmp['city'].idxmax()]
            result.append(result_row.product_id)
        else:
            dfTmp = df.copy()
            if not anyNone:
                already_chosen = findC(df, favCats, location)
            else:
                anyNone = True
            dfTmp = dfTmp[~dfTmp["product_id"].isin(already_chosen)]
            nearestCities = find5nearest(dfLogged, geolocation_unique_cities, location[0], cat)
            dfTmp = dfTmp[dfTmp["city"].isin(nearestCities)]
            dfTmp = dfTmp.loc[dfTmp['product_category_name'] == cat]
            dfTmp = dfTmp.groupby(["product_id"]).count().reset_index()
            result_row = dfTmp.loc[dfTmp['city'].idxmax()]
            result.append(result_row.product_id)
    return result
```

Example usage:

```python
findD(dfLogged, ["esporte_lazer", "brinquedos", "bebes"], ["tupa", "SP"])
```

Having almost zero information, we have a large pool of products to choose from. Let us limit these to further analysis only to those with the maximum average rating - all 5.0. It will also be displayed nicely and will inspire you in a positive way, encouraging you to buy even not necessarily including these products.

#### Second step

Let's merge it again with *reviews* - to filter only these having some nice comments.

```python
dfMain2 = dfMain2.merge(dfs['olist_order_reviews_dataset'], on="order_id", how="left")
dfMain2 = dfMain2.drop(["review_id", "review_score", "review_comment_title",
                        "review_creation_date", "review_answer_timestamp"], axis=1)
dfMain2 = dfMain2[dfMain2['review_comment_message'].notnull()]
dfMain2 = dfMain2[dfMain2['review_comment_message'].str.len() >= 20]
dfMain2 = dfMain2[dfMain2['review_comment_message'].str.len() <= 50]
dfMain2 = dfMain2.dropna()
dfMain2.sample(5)
dfMain2.shape
```

#### Third step

```python
dfMain2 = dfMain2[dfMain2['product_name_lenght'] >= 30]
dfMain2 = dfMain2[dfMain2['product_name_lenght'] <= 60]
dfMain2.sample(5)
dfMain2.shape
```

```python
dfMain2 = dfMain2.drop(["order_id", "seller_id", "zip_code_prefix", "lat", "lng",
                               "product_name_lenght", "product_description_lenght", "review_comment_message"], axis=1)
dfMain2.sample(5)
dfMain2.shape
```

#### Forth step

```python
dfMain2 = dfMain2[dfMain2["product_photos_qty"] >= 2]
dfMain2 = dfMain2.drop(["product_photos_qty"], axis=1)
dfMain2.sample(5)
dfMain2.shape
```

#### Five step

Let me take a look at the number of products "left in the pot".

```python
dfMain2countTmp = dfMain2[["product_id", "product_category_name"]].groupby(["product_id"]).count().reset_index()
dfMain2countTmp.sample(5)
dfMain2countTmp.shape
```

OK! So we have 2441 unique products; about 40% less than earlier. What about products, which were sold more than 5 times?

```python
dfMain2countTmp = dfMain2countTmp[dfMain2countTmp["product_category_name"] >= 5]
dfMain2countTmp.columns = ["product_id", "sales_count"]
dfMain2countTmp.sample(5)
dfMain2countTmp.shape
```

Great! Less than 100 products to choose from.

```python
dfMain2 = dfMain2.merge(dfMain2countTmp, on = "product_id")
dfMain2 = dfMain2.drop_duplicates()
dfMain2 = dfMain2.drop(["mean_score"], axis=1)
dfMain2.sample(5)
```

#### Sixth step

```python
dfMain2countTmp = dfMain2[["city", "product_id"]].groupby(["city"]).count().reset_index()
dfMain2countTmp = dfMain2countTmp.sort_values("product_id")
dfMain2countTmp.tail()
dfMain2countTmp.shape
```

For each city with >= 3 products, from the table I will remove the *floor(1/2)* of the products from that city that sold the least.

```python
popularCities = dfMain2countTmp[dfMain2countTmp["product_id"] >= 3]
popularCities["how_many_to_remove"] = popularCities["product_id"] // 2
popularCities
```

```python
isFirstIteration = True
for index, row in popularCities.iterrows():
    cityHere = row["city"]
    howManyToRemove = row['how_many_to_remove']
    dfMain2tmp = dfMain2[dfMain2["city"] == cityHere]
    dfMain2tmp = dfMain2tmp.sort_values("sales_count")
    dfMain2tmp = dfMain2tmp.iloc[howManyToRemove:]
    if isFirstIteration:
        dfMain2final = dfMain2tmp
        isFirstIteration = False
    else:
        dfMain2final = dfMain2final.append(dfMain2tmp)
dfMain2final.sample(5)
dfMain2final.shape
```

```python
dfMain2final = dfMain2final.append(dfMain2[~dfMain2["city"].isin(popularCities["city"])])
dfMain2final.sample(5)
dfMain2final.shape
```

```python
bestProductsNotLoggedIn = list(dfMain2final["product_id"])
bestProductsNotLoggedIn[1:10]
len(bestProductsNotLoggedIn)
```

### Side functions

```python
def predictKnown(favCats, location):
    listA = findA(dfLogged, favCats, location)
    result_list = [listA]
    
    listB = findB(dfLogged, favCats, location)
    listC = findC(dfLogged, favCats, location)
    listD = findD(dfLogged, favCats, location)
    BCD = (listB, listC, listD)
    
    for sublist in BCD:
        for item in sublist:
            result_list.append(item)
    
    return result_list
```

Example usage:

```python
predictKnown(["esporte_lazer", "brinquedos", "bebes"], ["tupa", "SP"])
```

```python
def predictUnknown():
    return random.sample(bestProductsNotLoggedIn, 10)
```

### Final function

```python
def predictTopProducts(favCats = None, location = None):
    if favCats is None or location is None:
        return predictUnknown()
    else:
        try:
            favCats = np.array(favCats)
            location = np.array(location)
        except Exception:
            print("Favorite categories and locations should be objects capable of being dumped into the numpy array")
        else:
            assert isinstance(favCats, np.ndarray)
            assert isinstance(location, np.ndarray)
            for i in range(3):
                assert isinstance(favCats[i], str), "Favorite categories should be strings"
            assert len(favCats) == 3, "Exactly three favorite categories needed"
            assert len(location) == 2, "Exactly two location strings [customer_city & customer_state] needed"
            assert isinstance(location[0], str) and isinstance(location[1], str), "Locations should be strings"
            return predictKnown(favCats, location)
```

Example usage:

```python
predictTopProducts(["esporte_lazer", "brinquedos", "bebes"], ["tupa", "SP"])
```
